```python
x = [1, 2, 3, 4]
y = [5, 4, 3, 2]

# create new figure
figure('demo')

# divide subplots into 2x3 grid
# select #1
subplot(231) 
plot(x, y)

# select #2
subplot(232)
bar(x, y)

# horizontal bar-charts
subplot(233)
barh(x, y)

# create stacked bar chars
subplot(234)
bar(x, y)

y1 = [7, 8, 5, 3]
bar(x, y1, bottom=y, color='r')

# box plot
subplot(235)
boxplot(x)

# scatter box
subplot(236)
scatter(x, y)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
y = np.cos(x)
y1 = np.sin(x)
plt.plot(x, y)
plt.plot(x, y1)
plt.title('Functions $\sin$ and $\cos$')
# set x & y limit
plt.xlim(-3.0, 3.0)
plt.ylim(-1.0, 1.0)

# format ticks at specific values
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],
           [r'$-1$', r'$0$', r'$+1$'])
plt.show()
```
